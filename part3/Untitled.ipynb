```python
import gmplot
import matplotlib.image as mpimg
gmap = gmplot.GoogleMapPlotter(40.088, -88.281, 16)
gmap.scatter(df["latitude"],df["longitude"], size=40, marker=False)
#Save to scatter.html
gmap.draw("scatter.html")

#Screenshot of website
scatter=mpimg.imread('scatter.png')
imgplot = plt.imshow(scatter)
plt.rcParams["figure.figsize"] = (15,15)
plt.title("Screenshot of Googlemap")
```
